```python
inequality_cols = [col for col in df.columns if col.startswith('Inequality in income')]
inequality_cols
```

```python
df[inequality_cols].isna().sum()
```

```python
df[inequality_cols] = df[inequality_cols].apply(
    lambda row: row.fillna(row.median()),
    axis=1
)
```

```python
df.head(10)
```

```python
df[inequality_cols].isna().sum()
```

```python
countries_all_missing = df[df[inequality_cols].isna().all(axis=1)]
countries_all_missing
```

```python
df_valid= df.dropna(subset=inequality_cols,inplace=False)
```

### Countries with Completely Missing Data

Some countries do not report income inequality values for any year between 2010 and 2021.
Since no historical information is available for these cases, median-based imputation is not possible.

Such countries were excluded from the analysis to avoid introducing artificial or misleading estimates.


```python
df_valid.isna().sum()
```

```python
df['UNDP Developing Regions'] = df['UNDP Developing Regions'].fillna('Not Classified')
```

```python
df['UNDP Developing Regions']
```

### Missing Categorical Data

The `UNDP Developing Regions` variable contains missing values for a subset of countries due to unavailable or unassigned UNDP classifications.
Rather than excluding these countries or inferring categories, missing values were labeled as **"Not Classified"** to preserve data completeness while maintaining transparency.


```python
df['UNDP Developing Regions'].value_counts()
```

```python
df_long = df.melt(
    id_vars=[
        'ISO3',
        'Country',
        'Continent',
        'Hemisphere',
        'Human Development Groups',
        'UNDP Developing Regions',
        'HDI Rank (2021)'
    ],
    value_vars=inequality_cols,
    var_name='year',
    value_name='income_inequality'
)
```

```python
df_long
```

```python
global_trend=df_long.groupby('year')['income_inequality'].mean().reset_index()
global_trend
```

```python
sns.barplot(data=global_trend, x='year', y='income_inequality',palette="viridis")
plt.xticks(rotation=90)
plt.figure(figsize=(10,6))
plt.show()
```

##Observations: Income Inequality Trend in India (20102021)

* Income inequality in India shows a gradual upward trend over the period 20102021, indicating a slow but persistent widening of income disparities.

* The increase is not abrupt, suggesting that inequality has risen incrementally rather than due to sudden economic shocks.

* From around 2015 onwards, income inequality appears to remain consistently higher than in the earlier years, pointing to a structural shift rather than short-term fluctuation.

* The relative stability of the trend in later years suggests that while inequality is increasing, it is doing so at a controlled but sustained pace.

* Overall, the trend indicates that economic growth during this period has not translated into more equitable income distribution, highlighting the need for targeted redistributive policies.

```python
india_trend=df_long[df_long['Country']=='India'].groupby('year')['income_inequality'].mean().reset_index()
india_trend
```

```python
global_trend=df_long.groupby('year')['income_inequality'].mean().reset_index()
global_trend
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=india_trend['year'],y= india_trend['income_inequality'],marker='o',label='India')

sns.lineplot(x=global_trend['year'],y=global_trend['income_inequality'],marker='o',linestyle='--',label='Global Average')
plt.title('Income Inequality Trend: India vs Global Average (20102021)')
```

## Observations: India vs Global Income Inequality (20102021)

* In 2010, Indias income inequality stood at approximately 19, while the global average was significantly higher at around 24, indicating a substantial initial gap between India and global inequality levels.

* Over the period, the global average inequality declined from about 24 to 22.8, reflecting gradual improvements in income distribution at the global level.

* In contrast, Indias inequality increased slightly from around 19 to approximately 19.3, showing little to no improvement over the same period.

* While the global trend suggests progress toward more equitable income distribution, Indias relatively flat trajectory indicates that similar gains have not been realized domestically.

* The narrowing gap between India and the global average is therefore driven primarily by global improvement rather than domestic reduction in inequality.

* This divergence highlights that, despite economic growth, India has not experienced proportional progress in reducing income inequality, underscoring the need for more inclusive growth strategies.

```python
sns.boxplot(data=df_long,x='Human Development Groups',y='income_inequality')
plt.title('Income Inequality by Human Development Group')
plt.xlabel('Human Development Group')
plt.ylabel('Income Inequality')
```

## Observations: Income Inequality by Human Development Group

* Countries in the Very High human development group exhibit the lowest median income inequality, indicating more equitable income distribution compared to other groups.

* Low, Medium, and High development groups show higher median inequality levels and a wider spread, suggesting greater disparities in income distribution.

* The Medium development group displays the largest variability in income inequality, with several extreme outliers, indicating that countries at similar development levels can experience vastly different inequality outcomes.

* The presence of high outliers in the Low and Medium groups suggests that economic growth alone does not guarantee equitable income distribution and may, in some cases, coincide with widening income gaps.

* The narrower interquartile range observed for Very High development countries reflects more consistent inequality levels, likely supported by stronger redistributive policies and social welfare systems.

* Overall, the plot indicates a negative relationship between human development level and income inequality, where higher development is generally associated with lower and more stable inequality.
